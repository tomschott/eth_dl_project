```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
# This notebook plots a heatmap with the expected error when quantizing a random matrix with fixed singular values and random uniform singular vectors.
# The expected error values were calculated with https://www.desmos.com/calculator/dhqpjjqsqw (may take a long time to calculate).

error_matrix = np.zeros((8,8))
error_matrix[0,0] = 0.256035503292
error_matrix[0,1] = 0.313242762339
error_matrix[0,2] = 0.320676126476
error_matrix[0,3] = 0.324363023657
error_matrix[0,4] = 0.325630077048
error_matrix[0,5] = 0.327711740948
error_matrix[0,6] = 0.328335071315
error_matrix[0,7] = 0.327767771316

error_matrix[1,1] = 0.278016106527
error_matrix[1,2] = 0.317682262762
error_matrix[1,3] = 0.323951071712
error_matrix[1,4] = 0.325299144449
error_matrix[1,5] = 0.326580405264
error_matrix[1,6] = 0.32742248564
error_matrix[1,7] = 0.328130030072

error_matrix[2,2] = 0.287991246616
error_matrix[2,3] = 0.320062239851
error_matrix[2,4] = 0.324151726309
error_matrix[2,5] = 0.325990721729
error_matrix[2,6] = 0.327246238421
error_matrix[2,7] = 0.328386763698

error_matrix[3,3] = 0.293959598615
error_matrix[3,4] = 0.321465389501
error_matrix[3,5] = 0.324973003658
error_matrix[3,6] = 0.326604777194
error_matrix[3,7] = 0.327516707046

error_matrix[4,4] = 0.297977633606
error_matrix[4,5] = 0.322374578977
error_matrix[4,6] = 0.325466982498
error_matrix[4,7] = 0.327051235224

error_matrix[5,5] = 0.300591956179
error_matrix[5,6] = 0.322714003937
error_matrix[5,7] = 0.325855524229

error_matrix[6,6] = 0.303073467649
error_matrix[6,7] = 0.323440409848

error_matrix[7,7] = 0.305008199997

for i in range(8):
  for j in range(i+1,8):
    error_matrix[j][i] = error_matrix[i][j]

error_matrix
```

```python
# Load expected error values into dataframe so we can give it to plotting library.
source = pd.DataFrame()
lamb_one = []
lamb_two = []
error = []
for i in range(1,9):
  for j in range(1,9):
    lamb_one.append(i)
    lamb_two.append(j)
    error.append(error_matrix[i-1][j-1])

source["位1"] = lamb_one
source["位2"] = lamb_two
source["Expected Quantification Error"] = error
source["error"] = source["Expected Quantification Error"]
```

```python
base = alt.Chart(source).encode(
    alt.X('位2:O'),
    alt.Y('位1:O'),
)

# Make heatmap.
heatmap = base.mark_rect().encode(
    alt.Color('Expected Quantification Error:Q')
        .scale(scheme='viridis')
        .title("Expected Quantification Error")
)

# Change color of the text depending on the background so it is always visible.
color = (
    alt.when(alt.datum.error > 0.306)
    .then(alt.value("black"))
    .otherwise(alt.value("white"))
)

# Make text to overlay on heatmap.
text = base.mark_text(baseline='middle').encode(
    alt.Text('Expected Quantification Error:Q', format=".3f"), color=color
)

# Draw the chart.
(heatmap + text).properties(width=300,height=300)
```
