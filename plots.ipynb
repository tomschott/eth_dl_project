```python
import matplotlib.pyplot as plt
import torch
import numpy as np

import json

```

```python
data = json.load(open('results/training_stats_avg.json', 'r'))
print(data.keys())
for k in data.keys():
    linestyles = ['solid', '--',]
    if k == 'AdamW':
        ls = linestyles[0]
    else:
        ls = linestyles[1]
    print(data[k].keys())
    #print(data[k]['L1_top_eigenvalues'])
    plt.semilogy(data[k]['L1_top_eigenvalues'],label=f'{k} L1', linestyle=ls)
    
    plt.semilogy(data[k]['L2_top_eigenvalues'], label=f'{k} L2', linestyle=ls)
    plt.semilogy(data[k]['L3_top_eigenvalues'], label=f'{k} L3', linestyle=ls)
    plt.semilogy(data[k]['L4_top_eigenvalues'], label=f'{k} L4', linestyle=ls)
plt.xlabel('Epochs')
plt.ylabel('L3 Top Eigenvalues')
plt.title('L3 Top Eigenvalues over Training')
plt.legend()
plt.show()
```
